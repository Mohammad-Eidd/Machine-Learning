```python
# Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
# Load Dataset

df = pd.read_csv("world_university_survey_dataset.csv")
```

```python
# Target Variable Distribution

df['overall_satisfaction'].value_counts()
```

```python
#  Satisfaction Count Plot

plt.figure(figsize=(8,5))
sns.countplot(x='overall_satisfaction', data=df)
plt.xticks(rotation=45)
plt.title("Distribution of Overall Satisfaction")
plt.show()
```

```python
#  Average Tuition by Country

country_tuition = df.groupby('country')['tuition_usd'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,6))
country_tuition.head(10).plot(kind='bar')
plt.title("Top 10 Countries by Average Tuition")
plt.ylabel("Average Tuition (USD)")
plt.show()
```

```python
#  Teaching Quality Rating vs Satisfaction

plt.figure(figsize=(8,5))
sns.boxplot(x='overall_satisfaction', y='teaching_quality_rating', data=df)
plt.xticks(rotation=45)
plt.title("Teaching Quality Rating vs Satisfaction Level")
plt.show()

```

```python
#  Tuition vs Teaching Quality (Scatter Plot)

plt.figure(figsize=(8,5))
sns.scatterplot(x='tuition_usd', y='teaching_quality_rating',
                hue='overall_satisfaction', data=df)
plt.title("Tuition vs Teaching Quality Rating")
plt.show()

```

```python
# Correlation Matrix Heatmap

plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```
